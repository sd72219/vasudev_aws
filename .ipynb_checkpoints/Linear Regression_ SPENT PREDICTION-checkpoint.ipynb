```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from statsmodels.stats.outliers_influence import variance_inflation_factor

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Ecommerce Customers")
df.head()
```

```python
df.drop(['Email', 'Address', 'Avatar'], axis=1, inplace=True)
```

```python
df.corr()
```

```python
plt.figure(figsize=(10,2))

sns.heatmap(df.corr().tail(1),annot=True, cmap='viridis')
```

```python
df.drop('Time on Website',axis=1, inplace=True)
```

### Model Training

```python
x = df.drop('Yearly Amount Spent',axis=1)
y = df['Yearly Amount Spent']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
```

```python
y_pred_test = linear_model.predict(x_test)
y_pred_train = linear_model.predict(x_train)
```

### Model Evalution

```python
mae_test = mean_absolute_error(y_test, y_pred_test)
mae_test
```

```python
mae_train = mean_absolute_error(y_train, y_pred_train)
mae_train
```

```python
mse_test = mean_squared_error(y_test, y_pred_test)
mse_test
```

```python
mse_train = mean_squared_error(y_train, y_pred_train)
mse_train
```

```python
score_test = r2_score(y_test, y_pred_test)
score_test
```

```python
score_train = r2_score(y_train, y_pred_train)
score_train
```

```python
y_pred_test1 = pd.DataFrame(y_pred_test)
y_test1 = y_test.reset_index(drop=True)
df_ = pd.concat([y_test1,y_pred_test1],axis=1)
df_.columns = ["y_test","y_pred_test"]
plt.figure(figsize=(15,10))
plt.plot(df_)
plt.legend(["ACTUAL VALUES" , "MODEL PREDICTION"])
```

```python
y_pred_train2 = pd.DataFrame(y_pred_train)
y_train2 = y_train.reset_index(drop=True)
df_ = pd.concat([y_train2,y_pred_train2],axis=1)
df_.columns = ["y_Train","y_pred_train"]
plt.figure(figsize=(15,10))
plt.plot(df_)
plt.legend(["ACTUAL VALUES" , "MODEL PREDICTION"])
```
